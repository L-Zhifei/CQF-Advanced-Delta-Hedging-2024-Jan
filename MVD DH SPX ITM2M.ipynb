```python
def combine_year_files(year_folder_path, year):
    # Initialize an empty list to store DataFrames
    dataframes = []

    # Iterate over each month (01 to 12)
    for month in range(1, 13):
        file_name = f"spx_eod_{year}{month:02d}.txt"
        file_path = os.path.join(year_folder_path, file_name)
        
        if os.path.exists(file_path):
            # Read the data file into a DataFrame
            df = pd.read_csv(file_path, delimiter=',')  
            dataframes.append(df)
        else:
            print(f"File {file_path} does not exist.")

    # Concatenate all DataFrames into one
    if dataframes:
        combined_df = pd.concat(dataframes, ignore_index=True)
        return combined_df
    else:
        return pd.DataFrame()  # Return an empty DataFrame if no files were read

# Example usage
year_folder_path = dir_path + spx_path + '/spx_eod_2010'  # Replace with your specific year folder path
year = 2010  # Specify the year
df_2010 = combine_year_files(year_folder_path, year)

```

# 0.2. Bucketing

```python
df_201023['c_delta'] = df_201023['c_delta'].apply(pd.to_numeric, errors='coerce')
```

```python
#bucketing by moneyness
def moneyness_bucketing(value):
    if 0.45 <= abs(value) <= 0.55:
        return 'ATM'
    elif 0 <= abs(value) <= 0.4:
        return 'OTM'
    elif 0.6 <= abs(value) <= 1:
        return 'ITM'
    else:
        return None

df_201023['moneyness_bucket'] = df_201023['c_delta'].apply(moneyness_bucketing)
```

```python
# bucketing by maturities
def days_to_maturity_bucketing(value):
    if 0<=value<=30:
        return '1M'
    elif 31<=value<=60:
        return '2M'
    elif 61<=value<=91:
        return '3M'
    elif 92<=value<=122:
        return '4M'
    elif 123<=value<=182:
        return '6M'
    elif 183<=value<=365:
        return '12M'
    elif value >= 366:
        return '12MM'
    else:
        return None

df_201023['maturity_bucket'] = df_201023['dte'].apply(days_to_maturity_bucketing)
```

```python
def option_pricing_call(df):
    S = df['underlying']
    K = df['strike']
    tau = df['dte']/256
    sigma = df['c_iv']
    r = df['rate']/100
    d1 = (np.log(S/K) + (r + (1/2) * sigma**2) * tau)/(sigma * np.sqrt(tau))
    d2 = d1 - sigma * np.sqrt(tau)
    value = S * norm.cdf(d1) - K * np.exp(-r * tau) * norm.cdf(d2)
    return value
```

```python
def rolling_regression(df, window_size):
    unique_dates = df['date'].unique()
    results = []

    for start_date in unique_dates: # ensure to shift by one day every time
        end_date = start_date + pd.Timedelta(days = window_size - 1)

        window_data = df[(df['date'] >= start_date) & (df['date']<= end_date)]

        if len(window_data) > 1:
            y = window_data['LHS']
            X = window_data[['RHS1','RHS2']]
            X = sm.add_constant(X)

            model = sm.OLS(y, X).fit()
            results.append({
                'start_date': start_date,
                'end_date': end_date,
                'a': model.params['const'],
                'b': model.params['RHS1'],
                'c': model.params['RHS2'],
                'rsquared': model.rsquared,
                'aic': model.aic,
                'bic': model.bic,
                'pvalues': model.pvalues,
                'tvalues': model.tvalues
            })
    return pd.DataFrame(results)
```

# Rolling Daily

```python
def rolling_quadratic_regression(df1, window_size_days=1095):  # ~3 years (365 * 3)
    min_date = df1['date'].min()
    max_date = df1['date'].max()
    window_size = pd.Timedelta(days=window_size_days)
    
    results = []

    # Start rolling from the first date after the initial 3 years
    start_date = min_date + window_size
    while start_date <= max_date:
        end_date = start_date
        window_start_date = start_date - window_size

        # Select data within the 3-year window
        window_data = df1[(df1['date'] > window_start_date) & (df1['date'] <= end_date)]
        
        if len(window_data) > 1:  # At least two data points needed for regression
            y = window_data['LHS']
            X = window_data[['RHS1', 'RHS2']]
            X = sm.add_constant(X)

            # Fit the quadratic regression model
            model = sm.OLS(y, X).fit()

            # Store the coefficients for the current date
            results.append({
                'date': end_date,
                'a': model.params['const'],
                'b': model.params['RHS1'],
                'c': model.params['RHS2']
            })
        
        # Move forward by one day
        start_date = start_date + pd.DateOffset(days=1)

    return pd.DataFrame(results)

df1M_ATM_calibrating = df1M_ATM_regression_noinf
df1M_ATM_calibrating = df1M_ATM_calibrating.rename(columns = {'date_x':'date'})

df1M_ATM_calibrating.head()

df1M_ATM_calibrated = rolling_quadratic_regression(df1M_ATM_calibrating)
```

## Rolling Monthly

```python
from pandas.tseries.offsets import MonthEnd

def rolling_quadratic_regression_monthly(df1, window_size_days=1095):  # ~3 years (365 * 3)
    min_date = df1['date'].min()
    max_date = df1['date'].max()
    window_size = pd.Timedelta(days=window_size_days)
    
    results = []

    # Start rolling from the first date after the initial 3 years
    start_date = min_date + window_size
    while start_date <= max_date:
        end_date = start_date
        window_start_date = start_date - window_size

        # Select data within the 3-year window
        window_data = df1[(df1['date'] > window_start_date) & (df1['date'] <= end_date)]
        
        if len(window_data) > 1:  # At least two data points needed for regression
            y = window_data['LHS']
            X = window_data[['RHS1', 'RHS2']]
            X = sm.add_constant(X)

            # Fit the quadratic regression model
            model = sm.OLS(y, X).fit()

            # Apply the coefficient to the next month
            next_month_start = end_date + pd.DateOffset(days=1)
            next_month_end = next_month_start + MonthEnd(1)

            days_in_next_month = pd.date_range(next_month_start, next_month_end)

            # Store the coefficients for the current date
            results.append({
                'date': end_date,
                'a': model.params['const'],
                'b': model.params['RHS1'],
                'c': model.params['RHS2']
            })
        
        # Move forward by one day
        start_date = start_date + pd.DateOffset(months=1)

    return pd.DataFrame(results)


df1M_ATM_calibrated_monthly = rolling_quadratic_regression_monthly(df1M_ATM_calibrating)
```
